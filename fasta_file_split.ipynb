```python
import os

input_fasta_file = 'aa_trajectories/T_trajectories.fasta'
output_directory = 'trajectories_psipred_new'

os.makedirs(output_directory, exist_ok=True)

try:
    # Initialize variables to store sequence data
    sequence_header = None
    sequence_data = []
    sequence_count = 0

    with open(input_fasta_file, 'r') as fasta_file:
        for line in fasta_file:
            line = line.strip()
            if line.startswith('>'):
                # If a new sequence header is encountered, write the previous sequence to a file
                if sequence_header:
                    sequence_count += 1
                    sequence_name = sequence_header[1:].split()[0]
                    output_file_name = os.path.join(output_directory, f'{sequence_name}.fasta')
                    with open(output_file_name, 'w') as output_file:
                        output_file.write(sequence_header + '\n')
                        output_file.write('\n'.join(sequence_data) + '\n')
                    sequence_data = []  # Reset 
                sequence_header = line
            else:
                sequence_data.append(line)

    # Write the last sequence to a file
    if sequence_header:
        sequence_count += 1
        sequence_name = sequence_header[1:].split()[0]
        output_file_name = os.path.join(output_directory, f'{sequence_name}.fasta')
        with open(output_file_name, 'w') as output_file:
            output_file.write(sequence_header + '\n')
            output_file.write('\n'.join(sequence_data) + '\n')

    print(f'Split {sequence_count} sequences into separate files in the "{output_directory}" directory.')

except FileNotFoundError:
    print(f"The file '{input_fasta_file}' was not found.")

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
import os

folder_path = 'trajectories_psipred_new'

file_list = os.listdir(folder_path)

# Process each file in the folder
for file_name in file_list:
    if file_name.endswith('.fasta') or file_name.endswith('.fa'):
        file_path = os.path.join(folder_path, file_name)
        with open(file_path, 'r') as file:
            lines = file.readlines()
        
        # Remove line breaks from sequence lines
        sequence_lines = [line.strip() if not line.startswith('>') else line for line in lines]
        
        # Write the modified content back to the file
        with open(file_path, 'w') as file:
            file.write(''.join(sequence_lines))
        
        print(f'Removed line breaks from {file_name} and saved.')

print('Processing complete.')
```

```python
import os

# Directory containing your FASTA files
input_directory = "trajectories_psipred_new/"
# Output merged FASTA file
output_file = "all_trajectories_merged.fasta"

# Open the output file for writing
with open(output_file, "w") as merged_fasta:
    # Iterate through files in the input directory
    for filename in os.listdir(input_directory):
        if filename.endswith(".fasta"):
            file_path = os.path.join(input_directory, filename)
            
            # Open each input FASTA file and append its contents to the output file
            with open(file_path, "r") as input_fasta:
                merged_fasta.write(input_fasta.read())
                merged_fasta.write("\n")  # Add a newline character to separate sequences

print(f"Successfully merged {len(os.listdir(input_directory))} files into {output_file}.")
```

```python
def count_fasta_sequences(file_path):
    with open(file_path, "r") as fasta_file:
        sequence_count = 0
        for line in fasta_file:
            if line.startswith(">"):
                sequence_count += 1
    return sequence_count

# Specify the path to your FASTA file
fasta_file_path = "trajectories_merged.fasta"

sequences_count = count_fasta_sequences(fasta_file_path)
print(f"The FASTA file contains {sequences_count} sequences.")
```
